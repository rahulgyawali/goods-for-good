```python
import pandas as pd
```

```python
usps_columns = ['State','PHYSICAL CITY', 'PHYSICAL ZIP','PHYSICAL DELV ADDR']

df_usps = pd.read_csv('../public/dataset/usps_zip.csv', usecols = usps_columns)

df_usps.rename(columns={
    'PHYSICAL CITY': 'city',
    'State': 'state',
    'PHYSICAL ZIP': 'zip',
    'PHYSICAL DELV ADDR': 'address'
}, inplace=True)

df_usps
```

```python
campaigns_columns = ['event_id','state', 'event_type','begin_date_time','episode_title','episode_narrative','event_narrative','cz_name']

df_campaigns = pd.read_csv('../public/dataset/details_c.csv', usecols = campaigns_columns)

df_campaigns
```

```python
water_events = ['Flood', 'Flash Flood','Coastal Flood', 'Heavy Rain', 'Storm Surge/Tide']
heat_events = ['Excessive Heat', 'Heat', 'Drought']
wind_events = ['High Wind', 'Strong Wind', 'Thunderstorm Wind', 'Dust Storm']
marine_events = ['Rip Current', 'High Surf', 'Marine Hail', 'Marine Thunderstorm Wind', 'Marine Tropical Depression', 'Marine Tropical Storm', 'Waterspout']
severe_weather_events = ['Hail', 'Lightning','Tornado', 'Funnel Cloud']
fire_events = ['Wildfire', 'Debris Flow']
tropical_cyclone_events = ['Tropical Depression', 'Tropical Storm']
```

```python
for i, campaign in df_campaigns.iterrows():
    event_type = campaign["event_type"]
    main_event_type=''
    image_folder = ''
    if event_type in water_events:
        main_event_type = 'Flooding and Water-Related'
        image_folder = 'Water'
    elif event_type in heat_events:
        main_event_type = 'Heat and Drought'
        image_folder = 'Drought'
    elif event_type in wind_events:
        main_event_type = 'Wind and Storm'
        image_folder = 'Infrastructure'
    elif event_type in marine_events:
        main_event_type = 'Coastal and Marine'
        image_folder = 'Water'
    elif event_type in severe_weather_events:
        main_event_type = 'Severe Weather'
        image_folder = 'Infrastructure'
    elif event_type in fire_events:
        main_event_type = 'Fire and Land Instability
        image_folder = 'Wildfire
    elif event_type in tropical_cyclone_events:
        main_event_type = 'Tropical Cyclones'
        image_folder = 'Water'
    else:
        main_event_type = 'Other'
        
```
